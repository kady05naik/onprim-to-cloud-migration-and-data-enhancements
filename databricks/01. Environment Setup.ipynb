### Create Exeternal Location

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS ext_location_databricks_tranformations
    URL 'abfss://catalog@onprimtocloudstorgaeacc.dfs.core.windows.net/'
    WITH (STORAGE CREDENTIAL `databricks-transformations`)
    COMMENT 'This is ext loc for catalog'


```

### 1. Create Catalog

```sql
CREATE CATALOG IF NOT EXISTS  catalog_dev
   MANAGED LOCATION 'abfss://catalog@onprimtocloudstorgaeacc.dfs.core.windows.net/' 
   COMMENT 'This is the catalog';
```

### 2. Select created Catalog

```sql
USE CATALOG catalog_dev;
SELECT CURRENT_CATALOG();
```
